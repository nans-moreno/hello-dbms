# Job 5 - GROUP BY : Grouper et comparer

## L'idée

Disons que tu veux calculer le salaire moyen PAR FONCTION. Pas le salaire moyen total, mais moyen des devs, moyen des managers, moyen des designers.

C'est ce que GROUP BY fait.

**Durée** : ~5 min

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE employes (
    id INTEGER PRIMARY KEY,
    nom TEXT,
    fonction TEXT,
    salaire INTEGER
)
''')

cursor.executemany('''
INSERT INTO employes (nom, fonction, salaire)
VALUES (?, ?, ?)
''', [
    ('Alice', 'Dev', 3000),
    ('Bob', 'Manager', 4000),
    ('Charlie', 'Dev', 3200),
    ('Diana', 'Design', 2800),
    ('Eve', 'Manager', 4200)
])

conn.commit()
```

## Combien de gens par fonction ?

```python
cursor.execute('SELECT fonction, COUNT(*) FROM employes GROUP BY fonction')
resultats = cursor.fetchall()

df = pd.DataFrame(resultats, columns=['Fonction', 'Nombre'])
print('Nombre d\'employés par fonction:')
print(df)
```

## Salaire moyen par fonction

```python
cursor.execute('SELECT fonction, AVG(salaire) FROM employes GROUP BY fonction')
resultats = cursor.fetchall()

df = pd.DataFrame(resultats, columns=['Fonction', 'Salaire moyen'])
print('Salaire moyen par fonction:')
print(df)
```

## Salaire total et nombre par fonction

```python
cursor.execute('''
SELECT fonction, COUNT(*) as nb, SUM(salaire) as total, AVG(salaire) as moyen
FROM employes
GROUP BY fonction
ORDER BY moyen DESC
''')
resultats = cursor.fetchall()

df = pd.DataFrame(resultats, columns=['Fonction', 'Nombre', 'Total', 'Moyen'])
print('Stats complètes par fonction:')
print(df)
```
